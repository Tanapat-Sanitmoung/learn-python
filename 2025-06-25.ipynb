```python
from ultralytics import SAM
import matplotlib.pyplot as plt
from PIL import Image

# Load the model
# model = SAM("mobile_sam.pt")
model = SAM("models/sam_b.pt")
```

```python
result = model.predict("oncepiece.png")
```

```python
result[0].save("oncepiece_seg.png")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

axes[0].imshow(Image.fromarray(result[0].orig_img))
axes[0].axis('off')
axes[1].imshow(Image.open('oncepiece_seg.png'))
axes[1].axis('off')
plt.tight_layout()
plt.show()
```
