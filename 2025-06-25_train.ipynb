```python
from ultralytics import YOLO

model = YOLO("models/yolo11n-seg.pt")
model.info()
```

```python
results = model.train(data="segmetn.v6i.yolov11/data.yaml", epochs=1, imgsz=640)
```

```python
trained_model = YOLO("runs/segment/train/weights/best.pt")

result = trained_model.predict("oncepiece.png")
```

```python
result[0].save("oncepiece_bubble.png")
```
